```python
import folium
import pandas as pd
import numpy as np
import pyproj
from folium import plugins
```

```python
line_final=pd.read_csv('line_final.csv', encoding='cp949')
line_final
```

```python
point_loc=pd.read_csv('최종 지역 거점.csv', encoding='UTF-8')
point_loc
```

```python
point_line=pd.read_csv('point_line.csv', encoding='UTF-8')
point_line
```

```python
from sklearn.preprocessing import MinMaxScaler

data = line_final[['경도', '위도']]

scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans

k = 25

model = KMeans(n_clusters = k, random_state = 10)

model.fit(data_scale)

line_final['cluster'] = model.fit_predict(data_scale)
```

```python
groupclust = line_final.groupby(['cluster'])
```

```python
clustmean = pd.DataFrame(groupclust['경도','위도'].mean())
clustmean
```

```python
latitude = 37.559819
longitude = 126.963895

m = folium.Map(location=[latitude, longitude],
               zoom_start=11, 
               width=750, 
               height=500
              )
m
```

```python
for en in range(len(clustmean['경도'])):
    folium.CircleMarker([clustmean['위도'][en], clustmean['경도'][en]],
                        fill = True,
                        color = 'red',
                        radius = 2,
                        fill_color = 'red').add_to( m )

m
```

# 1. 강남구

```python
gangnam = final_file.loc[final_file["주소"].str.contains('강남구')].reset_index(drop=True)
gangnam
```

```python
latitude = 37.493739
longitude = 127.065934

gangnam_map = folium.Map(location=[latitude, longitude],
               zoom_start=13, 
               width=750, 
               height=500
              )
gangnam_map
```

```python
gangnam_sub = line_final[line_final['역명'] == '대치']
gangnam_sub
```

```python
folium.RegularPolygonMarker([gangnam_sub['위도'], gangnam_sub['경도']],
                        fill = True,
                        color = 'darkblue',
                        radius = 10,
                        popup = gangnam_sub['역명'],
                        number_of_sides = 6).add_to( gangnam_map )

gangnam_map
```

```python
from sklearn.preprocessing import MinMaxScaler

data = gangnam[['경도', '위도']]

scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data)
```

```python
from sklearn.cluster import KMeans

k = 9

model = KMeans(n_clusters = k, random_state = 10)

model.fit(data_scale)

gangnam['cluster'] = model.fit_predict(data_scale)
```

```python
groupclust = gangnam.groupby(['cluster'])
```

```python
gananam_mean = pd.DataFrame(groupclust['경도','위도'].mean())
gananam_mean
```

```python
for en in range(len(gananam_mean['경도'])):
    folium.CircleMarker([gananam_mean['위도'][en], gananam_mean['경도'][en]],
                        fill = True,
                        color = 'red',
                        radius = 5,
                        fill_color = 'red').add_to( gangnam_map )

gangnam_map
```

```python
for en in range(len(gangnam['경도'])):
    folium.CircleMarker([gangnam['위도'][en], gangnam['경도'][en]],
                        fill = True,
                        color = 'black',
                        radius = 2,
                        fill_color = 'black',
                        popup = gangnam['대피시설명'][en]).add_to( gangnam_map )

gangnam_map
```
