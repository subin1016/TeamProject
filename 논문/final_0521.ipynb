```python
line_loca_drop = line_loca.drop_duplicates(subset=['역명'], keep= 'first')
line_loca_drop
```

```python
final = merge_line.drop(del_list,axis=0)
final
```

```python
final_ok=pd.read_csv('final_ok.csv', encoding='cp949')
final_ok
```

```python
final_ok1 = final_ok.drop(final_ok[final_ok['정거장층수']=='지상역사'].index)
final_ok2 = final_ok1.drop(final_ok1[final_ok1['정거장층수']=='지상'].index)
final_ok3 = final_ok2.drop(final_ok2[final_ok2['정거장층수']=='고가'].index)
okok = final_ok3[['역번호', '역명', '경도', '위도']].reset_index(drop=True)
okok
```

```python
line_concat = pd.concat([okok,line9])[['역번호','역명','경도','위도']].reset_index(drop=True)
line_concat
```

```python
drop_list = ['하남검단산', '하남시청','하남풍산','미사','모란','수진','신흥','단대오거리','남한산성입구','산성','철산']
```

```python
drop_index = []
for i in range(len(line_concat['역명'])):
    if line_concat['역명'][i] in drop_list:
        drop_index.append(i)
```

```python
drop_index
```

```python
line_final = line_concat.drop(drop_index,axis=0).reset_index(drop=True)
line_final
```

```python
line_final.to_csv('line_final.csv',encoding='cp949')
```

```python
for en in range(len(line_final['경도'])):
    folium.CircleMarker([line_final['위도'][en], line_final['경도'][en]],
                        fill = True,
                        color = 'black',
                        radius = 2,
                        fill_color = 'black').add_to( m )

m
```

```python
import requests
import json

# 서울 행정구역 json raw파일(githubcontent)
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)

m
```
