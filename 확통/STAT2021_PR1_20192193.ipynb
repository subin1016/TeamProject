### 1) 기본 마피아 (시민 5명, 마피아 1명)

```python
import random
win = []
for x in range(10000):    
    mafia = 1
    citizen = 5
    player = []
    
    # 마피아, 시민 배치
    for i in range(mafia): 
        player.append(1)
    for i in range(citizen):
        pos = random.randrange(0,mafia+i+1)
        player.insert(pos,2)
    people = len(player)
    
    # 게임 시작
    while(1):
        while(1): # 낮
            day = random.randrange(0,people)
            if player[day] > 0:
                player[day] = player[day]*(-1)
                break
                
        if 2 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break
        
        while(1): # 밤
            night = random.randrange(0,people)
            if player[night] != 1 and player[night] > 0:
                player[night] = player[night]*(-1)
                break
        
        if 2 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break
    
    win.append(winner)
    
print("마피아의 승률 :",round(win.count(1)/len(win),5), "\n시민의 승률 :", round(win.count(2)/len(win),5))
```

```python
import random
win = []
for x in range(10000):
    mafia = 2
    citizen = 7
    medic = 1
    player = []
    medicDie = 0

    # 마피아, 시민 배치
    for i in range(mafia): 
        player.append(1)
    for i in range(citizen):
        pos = random.randrange(0,mafia+i+1)
        player.insert(pos,2)
    for i in range(medic):
        pos = random.randrange(0,mafia+citizen+i+1)
        player.insert(pos,3) #메딕 배치
    people = len(player)

    # 게임 시작
    while(1):
        while(1): # 낮
            day = random.randrange(0,people)
            if player[day] > 0:
                if player[day] == 3:
                    medicDie += 1
                player[day] = player[day]*(-1)
                break

        if 2 not in player and 3 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break

        while(1): # 밤
            night = random.randrange(0,people)
            if player[night] != 1 and player[night] > 0:
                for i in range(medic-medicDie):
                    save = random.randrange(0,people) #메딕의 지목
                    while (player[save] < 0): 
                        save = random.randrange(0,people)
                    if night==save:
                        break
                if player[night] == 3:
                    medicDie += 1
                player[night] = player[night]*(-1)
                break
        if 2 not in player and 3 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break

    win.append(winner)

print("마피아의 승률 :",round(win.count(1)/len(win),5), "\n시민의 승률 :", round(win.count(2)/len(win),5))
```

```python
import random
win = []
for x in range(10000):    
    mafia = 2
    citizen = 7
    policeDie = 0
    find = -1
    player = []

    # 마피아, 시민 배치
    for i in range(mafia): 
        player.append(1)
    for i in range(citizen):
        pos = random.randrange(0,mafia+i+1)
        player.insert(pos,2)
    pos = random.randrange(0,mafia+citizen+1)
    player.insert(pos,4)
    people = len(player)

    # 게임 시작
    while(1):
        while(1): # 낮
            if policeDie == 0 and find != -1: # 경찰이 살아있으면서 마피아를 찾은 경우
                player[find] = player[find]*(-1)
                find = -1
                break
            day = random.randrange(0,people)
            if player[day] > 0:
                if player[day] == 4:
                    policeDie = 1
                player[day] = player[day]*(-1)
                break

        if 2 not in player and 4 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break

        while(1): # 밤
            night = random.randrange(0,people)
            if player[night] != 1 and player[night] > 0:
                if policeDie == 0:
                    check = random.randrange(0,people)
                    while(player[check] == 4 or player[check] < 0):
                        check = random.randrange(0,people)
                    if player[check] == 1:
                        find = check
                if player[night] == 4:
                    policeDie = 1
                player[night] = player[night]*(-1)
                break

        if 2 not in player and 4 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break
    
    win.append(winner)
    
print("마피아의 승률 :",round(win.count(1)/len(win),5), "\n시민의 승률 :", round(win.count(2)/len(win),5))
```

### 2) 투표 확률에 가중치를 줌 (5배 올려줌)

```python
import random
win = []
for x in range(10000):    
    mafia = 2
    citizen = 7
    policeDie = 0
    find = -1
    over = 0
    player = []
    checkList = []

    # 마피아, 시민 배치
    for i in range(mafia): 
        player.append(1)
    for i in range(citizen):
        pos = random.randrange(0,mafia+i+1)
        player.insert(pos,2)
    pos = random.randrange(0,mafia+citizen+1)
    player.insert(pos,4)
    people = len(player)
    weight = [1]*people

    # 게임 시작
    while(1):
        while(1): # 낮
            if policeDie == 0 and find != -1: # 경찰이 살아있으면서 마피아를 찾은 경우
                weight[find] = 5
                find = -1
            day = random.choices(range(0,people), weights = weight, k = 1)[0]
            if player[day] > 0:
                if player[day] == 4:
                    policeDie = 1
                player[day] = player[day]*(-1)
                break

        if 2 not in player and 4 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break

        while(1): # 밤
            night = random.randrange(0,people)
            if player[night] != 1 and player[night] > 0:
                if policeDie == 0:
                    check = random.randrange(0,people)
                    while(player[check] == 4 or player[check] < 0 or check in checkList):
                        check = random.randrange(0,people)
                        over += 1
                        if over > 1000:
                            break
                    checkList.append(check)
                    if player[check] == 1:
                        find = check
                if player[night] == 4:
                    policeDie = 1
                player[night] = player[night]*(-1)
                break

        if 2 not in player and 4 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break
    
    win.append(winner)
    
print("마피아의 승률 :",round(win.count(1)/len(win),5), "\n시민의 승률 :", round(win.count(2)/len(win),5))
```

```python
import random
win = []
for x in range(10000):    
    mafia = 3
    citizen = 3
    policeDie = 0
    medicDie = 0
    find = -1
    identifying = 0
    player = []

    # 마피아, 시민 배치
    for i in range(mafia): 
        player.append(1)
    for i in range(citizen):
        pos = random.randrange(0,mafia+i+1)
        player.insert(pos,2)
    pos = random.randrange(0,mafia+citizen+1)
    player.insert(pos,3)
    pos = random.randrange(0,mafia+citizen+2)
    player.insert(pos,4)
    people = len(player)

    # 게임 시작
    while(1):
        while(1): # 낮
            if policeDie == 0 and find != -1: # 경찰이 살아있으면서 마피아를 찾은 경우
                player[find] = player[find]*(-1)
                find = -1
                break
            day = random.randrange(0,people)
            if identifying == 1:
                while(player[day] < 0 or player[day] == 4):
                    day = random.randrange(0,people)
            if player[day] > 0:
                if player[day] == 4:
                    policeDie = 1
                if player[day] == 3:
                    medicDie = 1
                player[day] = player[day]*(-1)
                break

        if 2 not in player and 3 not in player and 4 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break

        while(1): # 밤
            night = random.randrange(0,people)
            if player[night] != 1 and player[night] > 0:
                if policeDie == 0:
                    check = random.randrange(0,people)
                    while(player[check] == 4 or player[check] < 0):
                        check = random.randrange(0,people)
                    if player[check] == 1:
                        find = check
                        identifying = 1

                if medicDie == 0:
                    save = random.randrange(0,people) #메딕의 지목
                    while (player[save] < 0): 
                        save = random.randrange(0,people)
                    if night==save:
                        break

                if player[night] == 4:
                    policeDie = 1
                if player[night] == 3:
                    medicDie = 1
                player[night] = player[night]*(-1)
                break

        if 2 not in player and 3 not in player and 4 not in player:
            winner = 1 # 마피아 승리
            break
        elif 1 not in player:
            winner = 2 # 시민 승리
            break

    win.append(winner)

print("마피아의 승률 :",round(win.count(1)/len(win),5), "\n시민의 승률 :", round(win.count(2)/len(win),5))
```
